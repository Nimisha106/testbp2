Import the module

```python
import networkx as nx
import matplotlib.pyplot as plt 
import pandas as pd
import numpy 
```

# Generate two undirected network G1 and G2

```python
#Generate a network G1
#Read xlsx file as dataframe
#You need to change the path to your own
data=pd.read_excel('/Users/dongjiaoge/Desktop/MA214/Week 20/lab5_1.xlsx', sheet_name='G1') 
#Create an empty network
G1 = nx.Graph()
#Add edges (the pair node1-node2 in the dataframe)
G1.add_edges_from([(row["node1"], row["node2"]) for idx, row in data.iterrows()])

nx.draw(G1, with_labels=True, node_color = 'cyan',font_weight='bold')
#plt.savefig('5_1.png')
```

```python
#Generate a network G2
#Read xlsx file as dataframe
#You need to change the path to your own
data=pd.read_excel('/Users/dongjiaoge/Desktop/MA214/Week 20/lab5_1.xlsx', sheet_name='G2') 
#Create an empty network
G2 = nx.Graph()
#Add edges (the pair node1-node2 in the dataframe)
G2.add_edges_from([(row["node1"], row["node2"]) for idx, row in data.iterrows()])

nx.draw(G2, with_labels=True, node_color = 'cyan',font_weight='bold')
#plt.savefig('5_2.png')
```

# Compute the density of G1 and G2

```python
#Compute the density of G1 and G2
density_G1=nx.density(G1)
density_G2=nx.density(G2)
print("The density of G1 is ", density_G1, ", and the density of G2 is ", density_G2)
print("The number of nodes in G1 is ", len(list(G1.nodes)), 
     "The number of edges in G1 is ", len(list(G1.edges)))
print("The number of nodes in G2 is ", len(list(G2.nodes)), 
     "The number of edges in G2 is ", len(list(G2.edges)))
n=len(list(G2.nodes))#n is the number of nodes
print("The maximumn possible number of edges of G1 and G2 is ", int(n*(n-1)/2))
```

# Compute degree of each node in G1 and G2

```python
#Compute the degree of each node of G1 and G2
degree_G1=G1.degree()
degree_G2=G2.degree()

print("The degrees of every node in G1 are ", degree_G1)
print("The degrees of every node in G2 are ", degree_G2)
```

# Compute the mean degree of G1 and G2

```python
#Compute mean degree of G1 and G2
n1=len(list(G1.nodes))#n1 is the number of nodes of G1
n2=len(list(G2.nodes))#n2 is the number of nodes of G2
#########
#Compute the mean degrees of G1 and G2
mean_degree_G1 = sum(dict(degree_G1).values())/n1 
mean_degree_G2 = sum(dict(degree_G2).values())/n2
#########
#Print out the mean degrees of G1 and G2
print("The mean degree of G1 is ", mean_degree_G1)
print("The mean degree of G2 is ", mean_degree_G2)
```

# Generate two directed networks G3 and G4

```python
#Generate a network G3
#Read xlsx file as dataframe
#You need to change the path to your own
data=pd.read_excel('/Users/dongjiaoge/Desktop/MA214/Week 20/lab5_1.xlsx', sheet_name='G3') 
#Create an empty network
G3 = nx.DiGraph()
#Add edges (the pair node1-node2 in the dataframe)
G3.add_edges_from([(row["node1"], row["node2"]) for idx, row in data.iterrows()])

nx.draw(G3, with_labels=True, node_color = 'cyan',font_weight='bold')
#plt.savefig('5_3.png')
```

```python
#Generate a network G4
#Read xlsx file as dataframe
#You need to change the path to your own
data=pd.read_excel('/Users/dongjiaoge/Desktop/MA214/Week 20/lab5_1.xlsx', sheet_name='G4') 
#Create an empty network
G4 = nx.DiGraph()
#Add edges (the pair node1-node2 in the dataframe)
G4.add_edges_from([(row["node1"], row["node2"]) for idx, row in data.iterrows()])

nx.draw(G4, with_labels=True, node_color = 'cyan',font_weight='bold')
#plt.savefig('5_4.png')
```

# Compute the density of G3 and G4

```python
#Compute the density of G3 and G4
density_G3=nx.density(G3)
density_G4=nx.density(G4)
print("The density of G3 is ", density_G3, ", and the density of G4 is ", density_G4)
print("The number of nodes in G3 is ", len(list(G3.nodes)), 
     "The number of edges in G3 is ", len(list(G3.edges)))
print("The number of nodes in G4 is ", len(list(G4.nodes)), 
     "The number of edges in G4 is ", len(list(G4.edges)))
n=len(list(G3.nodes))#n is the number of nodes
print("The maximumn possible numer of edges of G3 and G4 is ", int(2*n*(n-1)/2))
```

# Compute the in-degree and out-degree of each node in G3 and G4

```python
#Compute the in-degree and out-degree of each node of G3 and G4
in_degree_G3=G3.in_degree()
out_degree_G3=G3.out_degree()
in_degree_G4=G4.in_degree()
out_degree_G4=G4.out_degree()

print("The in-degrees of every node in G3 are ", in_degree_G3)
print("The out-degrees of every node in G3 are ", out_degree_G3)
print("The in-degrees of every node in G4 are ", in_degree_G4)
print("The out-degrees of every node in G4 are ", out_degree_G4)
```

# Compute the mean in and out degrees of G3 and G4

```python
#Compute mean in-degrees and out-degrees of G3 and G4
n3=len(list(G3.nodes))#n3 is the number of nodes of G3
n4=len(list(G4.nodes))#n4 is the number of nodes of G4
#########
#Compute the mean in-degrees and out-degrees of G3 and G4
mean_in_degree_G3 = sum(dict(in_degree_G3).values())/n3 
mean_out_degree_G3 = sum(dict(out_degree_G3).values())/n3 
mean_in_degree_G4 = sum(dict(in_degree_G4).values())/n4
mean_out_degree_G4 = sum(dict(out_degree_G4).values())/n4
#########
#Print out the mean in- and out- degrees of G3 and G4
print("The mean in degree of G3 is ", mean_in_degree_G3,
     "The mean out degree of G3 is ", mean_out_degree_G3)
print("The mean in degree of G4 is ", mean_in_degree_G4,
     "The mean out degree of G4 is ", mean_out_degree_G4)
```

# Get the degree distriution of G1 and G2

```python
#Degree distriutions of G1 and G2
degree_sequence_G1=sorted([d for n, d in G1.degree()], reverse=True)
degree_sequence_G2=sorted([d for n, d in G2.degree()], reverse=True)

#Obtain the unique values of the degrees, and the counts of each degree value
degree_unique_G1=numpy.unique(degree_sequence_G1, return_counts=True)
degree_unique_G2=numpy.unique(degree_sequence_G2, return_counts=True)

#Prepare the dataframes (degree and frequency) for visualizing the degree distriutions
#Column degree is from 1 to maximum value of degrees, 
# frequency gives the frequency with regard to each value in the degree column
degree_plot=list(range(0, max(degree_unique_G1[0])+1))
freq_plot=[0] * len(list(range(0, max(degree_unique_G1[0])+1)))
for i in degree_unique_G1[0]:
    freq_plot[i]=degree_unique_G1[-1][list(degree_unique_G1[0]).index(i)]
    
d = {'degree': degree_plot, 
     'frequency': freq_plot/sum(freq_plot)}
df_G1 = pd.DataFrame(data=d)


degree_plot=list(range(0, max(degree_unique_G2[0])+1))
freq_plot=[0] * len(list(range(0, max(degree_unique_G2[0])+1)))
for i in degree_unique_G2[0]:
    freq_plot[i]=degree_unique_G2[-1][list(degree_unique_G2[0]).index(i)]
    

d = {'degree': degree_plot, 
     'frequency': freq_plot/sum(freq_plot)}
df_G2 = pd.DataFrame(data=d)

#Plot the degree distriutions
fig, axes = plt.subplots(figsize=(7,10),nrows=2, ncols=1)

df_G1.plot.bar(x='degree', y='frequency', ax=axes[0])
axes[0].title.set_text("Degree distriution of G1")

df_G2.plot.bar(x='degree', y='frequency', ax=axes[1])
axes[1].title.set_text("Degree distriution of G2")

#plt.savefig('5_5.png')
```
